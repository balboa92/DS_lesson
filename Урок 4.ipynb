```python
t = np.linspace(0, 10, 54, dtype = 'float64')
print(t)
```

```python
f = np.cos(t)
print(f)
```

```python
plt.plot(t, f,linewidth=2.0, color = 'green')
plt.grid(True)
plt.axis([0.5,9.5,-2.5,2.5])
plt.title('График f(t)',fontsize = 24, fontname='Times New Roman' )
plt.xlabel('Значения t', color='red',fontsize = 20, fontname='Times New Roman')
plt.ylabel('Значения f',color='blue', fontsize = 20, fontname='Times New Roman')
plt.show()
```
